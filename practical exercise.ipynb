```python
# - 1. Write a program to drop rows with at least one Nan value (Null value)

import pandas as pd
data = {
    'a': [1,2,None,4],
    'b': [5,6,7,None]
}
df = pd.DataFrame(data)

df_cleaned = df.dropna()
print("Original data dataframe:")
print(df)
print('\n')
print("\nDataFrame after dropping rows with NaN values:")

print(df_cleaned)
```

```python
# - 2.Write a python program to draw multiple lines using plot() function

import matplotlib.pyplot as plt
x = [0,1,2,3,4]
y1 = [0,4,7,9,13]
y2 = [0,5,8,11,22]
plt.plot(x,y1)
plt.plot(x,y2)
plt.title("multiple lines using plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# - 3.Create a python program to display multiple types of charts using matplotlib package
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y1 = [1,2,3,4,5]
y2 = [2,4,6,7,9]
plt.plot(x,y1)
plt.title("Single line plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# - Multiline plot
plt.plot(x,y1)
plt.plot(x,y2)
plt.title("Multi line plot")
plt.xlabel("X")
plt.ylabel("y")
plt.show()

# - Bar chart
plt.bar(x,y1,color = "red")
plt.title("Bar chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# - Scatter plot
plt.scatter(x,y1)
plt.title("Scatter plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# - 5.Write a Python program to demonstrate various operations on Strings

txt = "hello welcome home"
x= txt.capitalize()
print(x)
a = txt.count("e")
print(a)
b = txt.endswith("e")
print(b)
c = txt.find("home")
print(c)
d = txt.isalpha() # - Check if all the characters in the text are letters and give output as True: 
print(d)
test = "12345"
e = test.isdigit()
print(e)
m = test.isupper() #- Check if all the characters in the text are in upper case:
print(m)
```
